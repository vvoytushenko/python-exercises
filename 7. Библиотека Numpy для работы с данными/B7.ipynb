```python
#Потренируемся извлекать из двумерного массива отдельные элементы и их срезы и производить над ними математические действия.
#Задания блока будем выполнять на примере массива, созданного кодом:

first_line = [x*y for x in range(2, 100, 6) for y in range (7, 1, -2)]
second_line = [x ** 0.5 for x in range(1000, 1101, 2)]
third_line = [x**2 for x in range(51)]

big_secret = np.array([first_line, second_line, third_line, second_line, first_line])
big_secret
```

```python
#9.3.1
#Чему равна сумма элементов последнего столбца массива? Ответ округлите до двух цифр после точки-разделителя.

print("%.2f" % big_secret.sum(axis=0)[-1])
```

```python
#9.3.2
#Выделите из каждой строки массива big_secret первые 5 элементов. 
#Чему равна сумма элементов главной диагонали получившейся матрицы? Округлите ответ до двух цифр после точки-разделителя.

b = big_secret[:, 0:5]
print("%.2f" % np.trace(b))
```

```python
#9.3.3
#Выделите из каждой строки массива big_secret последние 5 элементов. Чему равно произведение элементов главной диагонали
    #получившейся матрицы? Введите полученный результат без изменений и округлений, в качестве разделителя используйте точку.
    
b = big_secret[:, -5:]
print(b.diagonal().prod())
```

```python
#Важное отличие массивов NumPy от списков Python — при изменении среза или отдельных элементов все изменения касаются
    #не только самого среза, но и падают в исходный массив, даже если перед внесением изменений срез был сохранён в
    #виде отдельной переменной. Изначально NumPy проектировался для работы с большими массивами данных, 
    #поэтому при бесконтрольном копировании фрагментов массивов возникли бы проблемы с быстродействием и памятью.

#Создадим массив 4х6
my_array = np.random.randint(1, 100, (4, 6)) 
my_array
```

```python
#Проверим знания! Продолжим работать с массивом big_secret. Замените на 1 все элементы, у которых оба индекса нечётные,
    #и на -1 все элементы, у которых оба индекса чётные.
big_secret = np.array([first_line, second_line, third_line, second_line, first_line])
big_secret[1::1, 1::2] = 1
big_secret[0::2, 0::2] = -1
```

```python
students = np.array([[1, 135, 34, 4], [2, 160, 43, 5], [3, 163, 40, 4.3], [4, 147, 44, 5],
                    [5, 138, 41, 4.7], [6, 149, 54, 3.9], [7, 136, 39, 4.2], [8, 154, 48, 4.9], 
                    [9, 137, 35, 3.7], [10, 165, 60, 4.6]])
students
```

```python
#9.5.1
#Чему равно медианное значение массы тела школьников?

median_weight = np.median(students[:, -2])
median_weight
```

```python
#9.5.2
#На сколько среднее арифметическое массы тела школьников больше медианного значения для этого же показателя?
#Дайте ответ в килограммах с точностью до одной десятой кг, в качестве разделителя используйте точку.

avg_weight = np.mean(students[:, -2])
print(avg_weight)
print(avg_weight - median_weight)
```

```python
#9.5.3
#Из предложенных пар признаков в массиве students выберите ту, в которой наблюдается минимальная корреляция.

print(np.corrcoef(students[:, 1], students[:, 2]))
print(np.corrcoef(students[:, 1], students[:, 3]))
print(np.corrcoef(students[:, 2], students[:, 3]))
```

```python
#9.5.4
#Используя имеющийся набор данных о школьниках, найдите, чему равно стандартное отклонение их средних баллов:

np.std(students[:,3])
```

```python
#9.5.5
#Посчитайте, чему равна дисперсия значений веса школьников.
#Эту величину можно получить двумя способами: математически, используя определение дисперсии и известную функцию в блоке,
    #или самостоятельно найти подходящую функцию модуля NumPy.
    
np.var(students[:, 2])
```
