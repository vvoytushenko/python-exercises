```python
#1.8.1
#Чему равен средний возраст (Age), футболистов в наборе данных, округлённый до целого?
print('Age: ', football.Age.mean())

#Каково количество непустых строк в колонке Composure (Хладнокровие) набора данных о футболистах?
print('Composure: ', football.Composure.count())

#Каково в наборе данных о футболистах стандартное отклонение параметра коротких пасов (ShortPassing), округлённое до второго знака после запятой?
print('ShortPassing: ', football.ShortPassing.std())

#Какова сумма заработных плат за год (Wage) в наборе данных о футболистах?
print('Wage: ', football.Wage.sum())

#Какова минимальная стоимость футболиста?
print('Value: ', football.Value.min())
```

```python
# 1.9.1
#Какова средняя скорость (SprintSpeed) футболистов, зарплата (Wage) которых выше среднего? Ответ округлите до сотых.
print('#1.9.1: ', "{:.2f}".format(football[football.Wage > football.Wage.mean()].SprintSpeed.mean()))    
```

```python
#1.9.2
#Какова средняя скорость (SprintSpeed) футболистов, зарплата (Wage) которых ниже среднего? Ответ округлите до сотых.
print('#1.9.2: ', "{:.2f}".format(football[football.Wage < football.Wage.mean()].SprintSpeed.mean()))
```

```python
#1.9.3
#Какую позицию (Position) занимает футболист с самой высокой зарплатой (Wage)?
print('#1.9.3: ', football[football.Wage == football.Wage.max()].Position.max())
```

```python
#1.9.4
#Сколько пенальти (Penalties) забили бразильские (Nationality, Brazil) футболисты за период,
    #данные о котором представлены в датасете?
print('#1.9.4: ', football[football.Nationality == 'Brazil'].Penalties.sum())
```

```python
#1.9.5
#Укажите средний возраст (Age) игроков, у которых точность удара головой (HeadingAccuracy) > 50. Ответ округлите до сотых.
print('#1.9.5: ', "{:.2f}".format(football[football.HeadingAccuracy > 50].Age.mean()))
```

```python
#1.9.6
#Укажите возраст (Age) самого молодого игрока, у которого хладнокровие (Composure)
    #и реакция (Reactions) превышают 90% от максимального значения, представленного в датасете.
print('#1.9.6: ', football[(football.Composure > football.Composure.max()*0.9) 
                           & (football.Reactions > football.Reactions.max()*0.9) ].Age.min())
```

```python
#1.9.7
#Определите, на сколько средняя реакция (Reactions) самых взрослых игроков (т.е. игроков, чей возраст (Age)
    #равен максимальному) больше средней реакции самых молодых игроков. Ответ округлите до сотых.
print('#1.9.7: ', "{:.2f}".format(football[football.Age == football.Age.max()].Reactions.mean() 
                          - football[football.Age == football.Age.min()].Reactions.mean()))
```

```python
#1.9.8
#Из какой страны (Nationality) происходит больше всего игроков, чья стоимость (Value) превышает среднее значение?
            ##print('#1.9.8: ', football[football.Value > football.Value.mean()].Nationality.min())
ForNationality = football[(football.Value > football.Value.mean())]
NationalityCount = ForNationality['Nationality'].value_counts()
print('#1.9.8: ', NationalityCount.index[0])
```

```python
#1.9.9
#Во сколько раз средняя зарплата (Wage) голкипера (Position, GK) с максимальным значением показателя
    #«Рефлексы» (GKReflexes) выше средней зарплаты голкипера с максимальным значением
    #показателя «Владение мячом» (GKHandling)? Ответ округлите до сотых.
AvgGKWage1 = football[(football.Position == 'GK')&
        (football.GKReflexes == football.GKReflexes.max())].Wage.mean()
AvgGKWage2 = football[(football.Position == 'GK')&
        (football.GKHandling == football.GKHandling.max())].Wage.mean()
print('#1.9.9: ', "{:.2f}".format(AvgGKWage1/AvgGKWage2))
```

```python
#1.9.10
#Во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков (игроков с максимальным значением
    #показателя «Агрессивность» (Aggression)) выше силы удара игроков с минимальной агрессией? Ответ округлите до сотых.
MaxAggression = football[(football.Aggression == football.Aggression.max())].ShotPower.mean()
MinAggression = football[(football.Aggression == football.Aggression.min())].ShotPower.mean()
print('#1.9.10: ', "{:.2f}".format(MaxAggression/MinAggression))
```
