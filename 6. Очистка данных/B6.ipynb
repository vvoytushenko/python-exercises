```python
#7.11.7
#Теперь вернемся к нашему датафрейму log. Давайте преобразуем поле user_id, оставив там только идентификатор пользователя.
#Например, вместо "Запись пользователя № - user_974" должно остаться только "user_974".
#На месте записей с ошибками в user_id должна быть пустая строка "".
#Сделайте это через apply и новую функцию, которую вы создадите. Результат сохраните в log.

def user_update(old_user_id):
    if old_user_id == '#error':
        new_user_id = ''
        return new_user_id
    else:
        new_user_id = old_user_id.replace('Запись пользователя № - ', '')
        return new_user_id

log['user_id'] = log['user_id'].apply(lambda x: user_update(x))
log.head()
```
